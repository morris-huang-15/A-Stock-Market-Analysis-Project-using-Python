### Visualizing the "Total Traded" values

The Open Price Time Series Visualization makes Tesla look like its always been much more valuable as a company than GM and Ford. But to really understand this we would need to look at the total market cap of the company, not just the stock price. Unfortunately the current data doesn't have that information of total units of stock present. Instead, what can be done here is simple calcualtion to try to represent total money traded would be to multply the Volume column by the Open price. But this still isn't the actual Market Cap, its just a visual representation of the total amount of money being traded around using the time series. (e.g. 100 units of stock at \$10 each versus 100000 units of stock at $1 each)

```python
tesla['Total Traded'] = tesla['Open']*tesla['Volume']
ford['Total Traded'] = ford['Open']*ford['Volume']
gm['Total Traded'] = gm['Open']*gm['Volume']
```

Plot this "Total Traded" against the time index

```python
tesla['Total Traded'].plot(label='Tesla',figsize=(16,8))
gm['Total Traded'].plot(label='GM')
ford['Total Traded'].plot(label='Ford')
plt.legend();
```

### Visualizing "Moving Averages" (MA50 and MA200) 

In statistics, a moving average is a calculation used to analyze data points by creating a series of averages of different subsets of the full data set. In finance, a moving average (MA) is a stock indicator that is commonly used in technical analysis.

```python
gm['MA50'] = gm['Open'].rolling(50).mean()
gm['MA200'] = gm['Open'].rolling(200).mean()
gm[['Open','MA50','MA200']].plot(figsize=(16,8))
```

# Step3: Basic Financial Analysis


## Daily Percentage Change
Daily percentage change is defined by the following formula:

$ r_t = \frac{p_t}{p_{t-1}} -1$

This defines r_t (return at time t) as equal to the price at time t divided by the price at time t-1 (the previous day) minus 1. 

Create a new column for each dataframe called returns. This column will be calculated from the Close price column.

```python
tesla['returns'] = tesla['Close'].pct_change(1)
```

```python
ford['returns'] = ford['Close'].pct_change(1)
gm['returns'] = gm['Close'].pct_change(1)
```

```python
ford.head()
```

Now plot a histogram of each companies returns.

```python
ford['returns'].hist(bins=100)
```

```python
gm['returns'].hist(bins=100)
```

```python
tesla['returns'].hist(bins=100)
```

```python
tesla['returns'].hist(bins=100,label='Tesla',figsize=(10,8),alpha=0.5)
gm['returns'].hist(bins=100,label='GM',figsize=(10,8),alpha=0.5)
ford['returns'].hist(bins=100,label='Ford',figsize=(10,8),alpha=0.5)
plt.legend()
```

plotting a KDE instead of histograms for another view point

```python
tesla['returns'].plot(kind = 'kde',label = 'Tesla',figsize = (10,8))
ford['returns'].plot(kind = 'kde',label = 'Ford',figsize = (10,8))
gm['returns'].plot(kind = 'kde',label = 'GM',figsize = (10,8))
plt.legend()
```

creating some box plots comparing the returns

```python
box_df = pd.concat([tesla['returns'],ford['returns'],gm['returns']],axis=1)
box_df.columns = ['Tesla Ret','Ford Ret','GM Ret']
box_df.plot(kind='box',figsize=(8,11))
```

### Comparing Daily Returns between Stocks by plotting a Scatter Matrix

```python
scatter_matrix(box_df,figsize=(8,8),alpha=0.2,hist_kwds={'bins':100});
```

```python
box_df.plot(kind='scatter',x='Ford Ret',y='GM Ret',alpha=0.5,figsize=(8,8));
```

### Insight:
The scatter matrix clearly indicates that there is a strong positive correlation between Ford and GM's daily returns, whereas Tesla is completely different. It seems the market and investors treat Tesla a technology company rather than a car company like Ford or GM.

## Cumulative Daily Returns

The formula for a cumulative daily return is:

$ i_i = (1+r_t) * i_{t-1} $

Here we can see we are just multiplying our previous investment at i at t-1 by 1+our percent returns. Pandas makes this very simple to calculate with its cumprod() method. Using something in the following manner:

    df[daily_cumulative_return] = ( 1 + df[pct_daily_return] ).cumprod()
    

Create a cumulative daily return column for each car company's dataframe

```python
tesla['Cumulative Return'] = (1+tesla['returns']).cumprod()
```

```python
ford['Cumulative Return'] = (1+ford['returns']).cumprod()
gm['Cumulative Return'] = (1+gm['returns']).cumprod()
```

```python
tesla
```

Now plot the Cumulative Return columns against the time series index

```python
tesla['Cumulative Return'].plot(label='Tesla',figsize=(16,8),title='Cumulative Return')
ford['Cumulative Return'].plot(label='Ford')
gm['Cumulative Return'].plot(label='GM')
plt.legend()
```

### Insight:
Tesla stock showed the highest return for a $1 invested, compared to Ford which showed the lowest. An important actionable insight is Tesla stock has the best ROI and would be a good choice for long term investment.
